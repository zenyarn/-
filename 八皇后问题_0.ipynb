```python
""" GPT基于我的代码所给出的正确代码 """
class Solution:
    def backtrack(self, depth, board):
        if depth == self.size:  # 如果已经放置完所有行的皇后，记录解法
            self.ret.append(board[:])
            return

        # 构建可放置皇后的位置的选项集
        opt = [True] * self.size  # 标记这一行所有列的可用性
        for i in range(depth):
            line = board[i]
            # 找到之前行中的皇后，更新当前行的列和对角线的可用性
            for j in range(self.size):
                if line[j] == 'Q':
                    opt[j] = False  # 列冲突
                    if j + (depth - i) < self.size:
                        opt[j + (depth - i)] = False  # 主对角线冲突
                    if j - (depth - i) >= 0:
                        opt[j - (depth - i)] = False  # 次对角线冲突

        # 遍历每个可行的列位置
        for i in range(len(opt)):
            if opt[i]:
                new_line = ''.join(['Q' if i == j else '.' for j in range(self.size)])  # 构建新行

                depth += 1
                board.append(new_line)  # 将新行添加到棋盘
                self.backtrack(depth, board)  # 继续递归处理下一行
                board.pop()  # 回溯，移除当前行
                depth -= 1

    def solveNQueens(self, n: int) -> List[List[str]]:
        self.ret = []
        self.size = n

        # 开始回溯
        self.backtrack(0, [])

        return self.ret
```
